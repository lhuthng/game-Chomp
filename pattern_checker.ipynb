```python
from py_core import Board, Evidence
e = Evidence()
```

```python
def check(pattern, start=0, end=20):
    for i in range(start, end+1):
        b = Board(eval(pattern.replace('n', str(i))))
        if e.get(b): return False
    return True

def check_ideal(pattern, start, end=20):
    for i in range(start, end+1):
        translated = str(eval('('+pattern.replace('n', str(i))[1:-1]+')'))
        b = Board.get_board('<'+translated[1:-1]+'>')
        if e.get(b): 
            print("Error on", b)
            return False
    return True

def check_ideal_2(pattern, start=1, end=20, req=None, init=None):
    if not req:
        req = lambda n, m: m < n
    if not init:
        init = lambda _: 0
    for i in range(start, end+1):
        j = init(i)
        while req(i, j):
            translated = str(eval('('+pattern.replace('n', str(i)).replace('m', str(j))[1:-1]+')'))
            b = Board.get_board('<'+translated[1:-1]+'>')
            if e.get(b): 
                return False
            j += 1
    return True
```

```python
get_mex = lambda p, q: e.get_mex(Board([p, q]))
def get_mex_2(p, q):
    if q % 2 == p % 2:
        return p + (q - 1) // 2
    if 2 * q <= p:
        return p - (q+1)//2 - 1
    else:
        return 3*((p-q-1) // 2)
```
